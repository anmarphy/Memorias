## 3. Joins

```python
carpeta="C:/Users/Andrea/Desktop/python-ml-course-master/datasets/athletes"
medallas="Medals.csv"
paises="Athelete_Country_Map.csv"
atletas="Athelete_Sports_Map.csv"
```

```python
med=pd.read_csv(carpeta+"/"+medallas, encoding="ISO-8859-1")
med.head()
```

```python
a=med["Athlete"].unique().tolist()
len(a)
```

```python
pais=pd.read_csv(carpeta+"/"+paises, encoding="ISO-8859-1")
pais.head(n=2)
```

#### Inner Join

```python
mer_inner=pd.merge(left=med, right=pais, 
                  how="inner", left_on="Athlete", right_on="Athlete")
```

```python
mer_inner.tail(n=2)
```

#### Left Join

```python
mer_left=pd.merge(left=med, right=pais, 
                  how="left", left_on="Athlete", right_on="Athlete")
mer_left.tail(n=2)
```
